**DATA STANDARDIZATION**

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
print(dataset)
```

```python
cancer=pd.DataFrame(dataset.data, columns=dataset.feature_names)
print(cancer.head())
```

```python
cancer.shape
```

```python
x=cancer
y=dataset.target
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3,test_size=0.2)
```

**standardising data**

```python
print(dataset.data.std())
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train_standard=scaler.transform(x_train)
print(x_train_standard)
```

```python
x_test_standard=scaler.transform(x_test)
```

```python
print(x_train_standard.std())
```

```python
print(x_test_standard.std())
```

**HANDLING IMBALANCED DATASET**

```python
import pandas as pd
import numpy as np
```

```python
ccd=pd.read_csv('/content/credit_data.csv')
ccd.head()
```

```python
ccd['Class'].value_counts()
```

```python
legit = ccd[ccd.Class==0]
fraud = ccd[ccd.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**Undersampling**

```python
legit_sample=legit.sample(n=56)
legit_sample.shape
```

```python
nccd=pd.concat([legit_sample,fraud],axis=0)
nccd.shape
nccd['Class'].value_counts()
```
